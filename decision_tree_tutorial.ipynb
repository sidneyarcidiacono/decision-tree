```
# Import our iris dataset from sklearn
from sklearn.datasets import load_iris
# Import our decision tree model from sklearn
from sklearn import tree
# Import pandas
import pandas as pd
# Import matplotlib to fix the size of our tree plot
import matplotlib.pyplot as plt


# Load our dataset
iris = load_iris()
# Convert to DataFrame so we can see our data more clearly
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
# Create a new column with our labels
iris_df["Class"] = iris.target
iris_df

# Create x and y for sklearn to work with
# X is our features
# y is our labels (our type of iris)
X = iris_df.drop("Class", axis=1)
y = iris_df["Class"]

# Set our classifier (clf) equal to sklearn's DecisionTreeClassifier
# Call fit to train
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

# Set our figure size
plt.figure(figsize=(12, 12))

# Show our decision tree! 
tree.plot_tree(clf)
```
